### Sampling Distributions and Standard Errors

The sampling distribution of a sample statistic is a theoretical distribution that describes all possible values of a sample statistic from random samples of the same size, taken from the same population.

In "real-life" research, only one sample from each population under study will be taken.

### Estimating Sampling Distribution Characteristics from Single Samples of Data

#### The Central Limit Theorem (CLT)

There is a mathematical theorem that generalizes these properties called the Central Limit Theorem (CLT).

Basically, the CLT states the (theoretical) sampling distribution of a sample statistic will:
- Be approximately normal;
- Have average of the true, population-level value being estimated;
- Have variability that is a function of the variation of individual values in the population (sd) and the size of the sample the statistic is based upon: This variability in sample statistics across multiples samples of the same size is called the $\textbf{standard error}$.

For example, having a normal distribution with $\mu$ = 300, $\sigma = 30$ and sample size $n = 10$. The standard deviation for the sample is given by $\sigma_{x} = \frac{\sigma}{\sqrt(n)}$.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
from matplotlib.patches import Polygon
import math

n = 10
mu = 300
sigma = 30
sigma = sigma/math.sqrt(n)

fig, ax = plt.subplots()
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
ax.plot(x, stats.norm.pdf(x, mu, sigma))
ax.set_ylim(bottom=0)

a = mu - sigma
b = mu + sigma

plt.show()
```

But if the sample size is greater? Suppose $n = 100$. We would have a more closed normal distribution as variation decreases.

```python
n = 100
mu = 300
sigma = 30
sigma = sigma/math.sqrt(n)

fig, ax = plt.subplots()
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
ax.plot(x, stats.norm.pdf(x, mu, sigma))
ax.set_ylim(bottom=0)

a = mu - sigma
b = mu + sigma

plt.show()
```

### Confidence Intervals for Single Population Parameters

95% confidence intervals for a population mean $\mu$, based on a data in random sample taken from the population, can be constructed by:

$$\bar{x} \pm 2\widehat{\text{SE}}(\bar{x}), \qquad \text{Where} \;\widehat{\text{SE}}(\bar{x}) = \frac{s}{\sqrt{n}}$$

The standard error of the sample mean quantifies the variation in sample means across random samples of the same size (from the same population).

The level of confidence can be changed by adjusting the number of standard errors subtracted from and added to the sample mean.

For sample proportion:

$$\widehat{\text{SE}}(\hat{p}) = \sqrt{\frac{\hat{p}\times (1 - \hat{p})}{n}}$$

For incidence rate:

$$\widehat{\text{SE}}(\widehat{IR}) = \frac{\sqrt{\text{# events}}}{\text{Total Follow-up Time}}$$

The 95% (or other percent, if other level intervals are used) refers to how often the approach to creating a 95% CI "works" in general.

In other words, for 95% of samples randomly selected from a population, 95% confidence interval created from the sample will contain the true value of the parameter (mean, proportion, incidence rate, etc.) of interest.

#### Note about the Central Limit Theorem in general

The CLT requires "enough data" to kick in and the sampling distribution to be normal
- For example, with sample means, a sample size cutoff used is n = 60

Why does CLT require a larger sample? 
- Sample needs to be large enough so that influence of any single sample value is relatively small

"Large enough" varies depending on data type
- The good news is that computers can compute correct 95% (and o ther level) CIs regardless of the sample size, and the interpretation is the same.

So, the CLT tells us the following:
- When taking a random sample of size n from a population, the theoretical sampling distribution of a sample mean across all possible random samples of size n is a normal distribution.

#### "Correcting" for Smaller Samples with the t-distribution

If we have a smaller sample size, we will have to go out more than two standard errors (SEs) to achieve 95% confidence.

How much bigger the 2 depends on the sample size.

You can look up the correct number in a "t-table" or "t-distribution", with n-1 degrees of freedom.

How many standard errors we need to go depends on the degrees of freedom - this is linked to sample size.

The appropriate degrees of freedom are n - 1.

Using 95% CI with t-distribution:

$$\bar{x} \pm t_{0.95, n - 1}\times \widehat{\text{SE}}(\bar{x})$$


### Confidence Intervals for Population Comparison Measures

The sampling behavior of differences in quantities that both have approximately normal sampling distributions (means, proportions) have an approximately normal sampling distribution.

The sampling distributions of ratios (odds, risk, IR) on $\textit{ln}$ is approximately normal

The approach to getting 95% CIs for the above quantities at the population level will be:

$$(\text{estimate}) \pm 2\widehat{\text{SE}}(\text{estimate})$$

#### Confidence Intervals for Comparing Means of Continuous Outcomes Between Two Populations

95% (and other level CIs) can relatively easily be estimated for mean differences between two populations for both paired and unpaired study designs

For paired studies with paired sample data, the two samples can be transformed into a single sample of differences
- The standard error of the mean difference is a function of the standard error of the differences and the number of pairs

For unpaired studies, the two samples cannot be combined
- The standard error of the mean difference is a function of the standard error of the values in each of the two groups being compared

### Two-Group Hypothesis Testing: The General Concept and Comparing Means

Confidence intervals and hypothesis testing are two complementary ways of addressing uncertainty in sample-based comparisons when making statements about the unknown population comparisons

Both methos operate on the principle that for most random sample-based studies, the sample results should be "close" to the truth.

Paired: when two samples drawn from populations under study are linked systematically.

Unpaired: when the two samples are drawn from two independent, unliked populations.

#### Paired t-Test Approach

Translate the distance into a p-value and make a decision (p-value based on the normal distribution, or the t-distribution with smaller samples: computer will handle the details)

The p-value measures the chance of getting the study results (or something even less likely, i.e, more extreme) when the samples are assumed to have come from populations with the same Means

The p-value is invariant to the direction of comparison

#### Unpaired t-Test Approach

Translate the distance into a p-value and make a decision (p-value based on the normal distribution, or the t-distribution with smaller samples: computer will handle the details).

The p-value measures the chance of getting the study results (or something even less likely, i.e., more extreme) when the samples are assumed to have come from populations with the same means.

The p-value (this a called a two-sided p-value, more to come) is invariant to the direction of comparison.

The only difference between the paired and unpaired (two-sample) t-test is how the standard error of the sample mean difference is computed.

#### p-Value

The p-value alone can only indicate whether the study results were likely due to (random sampling), chance or not if there is no difference in the measure being compared between populations.
- p-Values should not generacilly not be reported alone, without the estimated mean difference and 95% CI endpoints

Not rejecting the null hypothesis is not equivalent to accepting the null hypothesis as the truth

#### Hypothesis Testing: The General Approach

State the null $H_{0}$ and alternative $H_{A}$ hypotheses

Assuming the null, and measure the difference between the study based estimate of the quantity of interest, and the assumed value under hypothesis in standard errors

Convert the standardized distance measure into a p-value
- The p-value is the probability of getting a result as extreme or more extreme (as unlikely or more unlikely) then the study results if the null is true

Make a decision to "reject" or "fail to reject" the null hypothesis based on the p-value.

### Hypothesis Testing: Comparing Proportions and Incidence Rates Between Two Populations

#### The "z-Test" Approach

The two-sample z-test provides a method for getting a p-value for testing two competing hypotheses about the true proportions of a binary outcome between two Populations

$$H_{0}: p_{1} = p_{2}$$
$$H_{A}: p_{1} \neq p_{2}$$

As such, only one p-value (and hence, one hypothesis test) is needed for all measures of association comparing binary outcomes between two Populations.

The test is performed using the observed difference in proportions (risk difference: $\hat{p}_{1} - \hat{p}_{2}$) and its estimated standard error ($\widehat{SE}(\hat{p}_{1} - \hat{p}_{2}$)).

#### Chi-Square and Fisher's Exact Test Approaches

The two-sample z-test and chi-square test give exactly the same result; however, the chi-quare is usually what is referred to in the literature.

Fisher's exact test is a computer based test - its results will usually align with the other two tests, but the resulting p-values may differ in slightly smaller samples.

Any of the three are generally appropriate for comparing proportions between two populations, and the resulting p-values are interpreted the same way.

#### Comparing Incidence Rates Between Two Populations

Both the two-sample z-test and the log rank can be used to test competing null and alternative hypothesis about time-to-event data.

The log rank is most commonly presented in the literature, but the two-sample z-test is a nice, easy-to-implement-by-hand approach that is very similar in its approach to the two-sample t-test for comparing means and the two-sample z-test for comparing proportions.

Because of slightly different mechanics, the p-values from both tests may differ slightly in value

Both tests use the same logic as all other hypothesis tests we've seen

The log rank test can be extended to compare survival curves between more than two populations with one test

### Comparing Means Between More Than Two Populations: Analysis of Variance (ANOVA)

Analysis of Variance (ANOVA) can be used to test for population level differences in means between two or more populations with one test.

The competing hypotheses are:

$$H_{0}: \mu_{1} = \mu_{2} = \cdots \mu_{k}$$
$$H_{A}: \text{at least two populations have different}\; \mu s$$

A significant p-value (<0.05) means that at least two populations have different means, but this result does not identify the specific populations that differ.
